5.1  Please combine **image crop, color shift, rotation and perspective transform** together to complete a data augmentation script.
   Your code need to be completed in Python/C++ in .py or .cpp file with comments and readme file to indicate how to use.

```python
import cv2
import random
import numpy as np
from matplotlib import pyplot as plt
```

```python
def image_crop(): # you code here
    img_origin1 = cv2.imread('lenna.jpg',1)  # 0：黑白图(二维灰度图)， 1彩图（三维彩图）
    print(img_origin1.shape)  # 图片大小
    img = img_origin1[100:200,200:300]  #  切分图片  
    plt.figure(figsize=(4,4))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # 彩图
    plt.show()  # 显示图片
    
image_crop()
```

```python
def color_shift(): # you code here
    img_origin1 = cv2.imread('dark.jpg',1)  # 0：黑白图(二维灰度图)， 1彩图（三维彩图）
    B,G,R = cv2.split(img_origin1)
    B = 255 - B
    G = 255 - G
    R = 255 - R
    
    img = cv2.merge((B,G,R))
    
    plt.subplot(121)
#     plt.figure(figsize=(4,4))
    plt.imshow(cv2.cvtColor(img_origin1, cv2.COLOR_BGR2RGB))  # 彩图
    
    plt.subplot(122)
#     plt.figure(figsize=(4,4))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # 彩图
    
    plt.show()  # 显示图片

color_shift()
```

```python
def rotation(): # you code here
    img_origin1 = cv2.imread('lenna.jpg',1)
    image = cv2.rotate(img_origin1, rotateCode = cv2.ROTATE_90_CLOCKWISE)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # 彩图
    plt.show()  # 显示图片
rotation()
```

```python
def perspective_transform(): # you code here
    pt1 = np.float32([[0,0],[0,500],[500,0],[500,500]]) # 原始点
    pt2 = np.float32([[0,0],[0,400],[300,0],[450,550]])  # 投射后点
    img_origin1 = cv2.imread('lenna.jpg',1)
    M = cv2.getPerspectiveTransform(pt1, pt2)  # 获取投射点
    image_warp = cv2.warpPerspective(img_origin1, M, (500,500))
    plt.imshow(cv2.cvtColor(image_warp, cv2.COLOR_BGR2RGB))  # 彩图
    plt.show()  # 显示图片
    
perspective_transform()
```
